## Korelasi antara satu variabel dengan variabel lainnya

```python
cols = list(data_frame)[2:]
cols
```

```python
def sort_correlation(column_list, reference):
    dict = {}
    for item in column_list:
        dict[item] = data_frame[reference].corr(data_frame[item])
    unsorted_corr = pd.DataFrame.from_dict(dict, orient='index',columns=['Correlation'])
    return unsorted_corr.sort_values(by='Correlation', ascending=False )
```

```python
# Untuk contoh di bawah yang kita gunakan sebagai referensi untuk pencarian korelasi adalah column outgoing.
# Tentunya kita bisa ganti dengan column lainnya misalnya column incoming
sorted_corr = sort_correlation(cols, "Outgoing")
sorted_corr.index.name = 'Factor'
sorted_corr.head()
```

```python

%matplotlib inline
plt.rcParams["figure.figsize"] = [25, 10]
sorted_corr.plot.bar();
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = [25, 30]
sorted_corr.plot.barh();
```

```python
#list(sorted_corr.index)
```

```python
alt.Chart(sorted_corr.reset_index()).mark_bar().encode(
    alt.Y('Factor', sort=list(sorted_corr.index)),
    x='Correlation:Q'
).properties(height=800, width=700)
```

## Perbandingan *Year-on-Year*

```python
def diff_yoy(month):
    df_filtered = data_frame[data_frame['Month (number)'] == month].set_index('Year')
    col = list(df_filtered)[:5]
    col.pop(0)
    return df_filtered[col].diff().reset_index()
```

```python
# data_frame.head()
diff_yoy(3).describe()
```

```python
# untuk contoh di bawah kita pilih perbandingannya di bulan ke 3 dari setiap tahun
%matplotlib inline
plt.rcParams["figure.figsize"] = [25, 15]
diff_mpl = diff_yoy(3).set_index('Year')
diff_mpl.plot.bar()
```

```python
# untuk contoh di bawah kita pilih perbandingannya di bulan ke 3 dari setiap tahun
alt.Chart(diff_yoy(3).melt('Year')).mark_bar().encode(
    alt.X('variable:N', axis=alt.Axis(title='')),
    alt.Y('value:Q', axis=alt.Axis(title='Delta', grid=False)),
    color=alt.Color('variable:N'),
    column='Year:O'
).configure_view(
    stroke='transparent'
).properties(height=500, width=100)
```

## Perbandingan *Month-on-Month*

```python
def diff_mom(year):
    df_filtered = data_frame[data_frame['Year'] == year].set_index('Month (number)')
    col = list(df_filtered)[:5]
    col.pop(0)
    return df_filtered[col].diff().reset_index()
```

```python
diff_mom(2018).describe()
```

```python
# untuk contoh di bawah kita pilih perbandingan MoM di tahun 2018
%matplotlib inline
plt.rcParams["figure.figsize"] = [25, 15]
diff_mpl = diff_mom(2018).set_index('Month (number)')
diff_mpl.plot.bar()
```

```python
# untuk contoh di bawah kita pilih perbandingan MoM di tahun 2018
alt.Chart(diff_mom(2018).melt('Month (number)')).mark_bar().encode(
    alt.X('variable:N', axis=alt.Axis(title='')),
    alt.Y('value:Q', axis=alt.Axis(title='Delta', grid=False)),
    color=alt.Color('variable:N'),
    column='Month (number):O'
).configure_view(
    stroke='transparent'
).properties(height=500, width=100)
```

## Analisa Trend

```python
date_time = ['Year', 'Month (number)']
dt = data_frame[date_time].assign(Day=28)
dt = dt.rename(columns={'Month (number)':'Month'})
dt = pd.to_datetime(dt)
dt = pd.DataFrame(dt)
dt = dt.rename(columns={0:'Time'})
dt.head()
```

```python
data_frame.index.name = 'key'
dt.index.name = 'key'
df_trend = data_frame.join(dt).set_index('Time')
cols = list(data_frame)[2:6]
df_trend = df_trend[cols]
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = [25, 15]
display = ['Incoming','Outgoing']
df_trend[display].plot.line(colormap='rainbow')
```

```python
display = ['Incoming','Outgoing']
data = df_trend[display].reset_index().melt('Time')
alt.Chart(data).mark_line().encode(
    alt.X('Time:T', axis=alt.Axis(title='Time')),
    alt.Y('value:Q', axis=alt.Axis(title='Amount', grid=False)),
    color=alt.Color('variable:N'),
).properties(height=600, width=800)
```

```python
df_trend['In-Out'] = df_trend['Incoming'] - df_trend['Outgoing']
```

```python
display = ['In-Out']
data = df_trend[display].reset_index().melt('Time')
alt.Chart(data).mark_line().encode(
    alt.X('Time:T', axis=alt.Axis(title='')),
    alt.Y('value:Q', axis=alt.Axis(title='Amount', grid=False)),
    color=alt.Color('variable:N'),
).properties(height=600, width=800)
```

### Penggunaan teknik interpolasi

```python
from scipy.interpolate import interp1d
import numpy as np
df_trend_reindexed = df_trend.reindex(pd.date_range(start=df_trend.index.min(), end=df_trend.index.max(), freq='1D'))
df_trend_reindexed.index.name = 'Time'
```

```python
df_trend_reindexed = df_trend_reindexed.interpolate(method='cubic')
```
